## 6.1 Why Do We Need Conditional Statements?<br>
Imagine writing a program that checks whether a person can vote.<br>
The rule is simple.
```text
Age ≥ 18
↓
Eligible
```
Otherwise,
```text
Not Eligible
```

Without conditional statements, Python would execute every statement one after another.<br>
Example:
```python
print("Eligible to Vote")
print("Not Eligible to Vote")
```
Output
```text
Eligible to Vote
Not Eligible to Vote
```
Obviously, this is incorrect because only **one** of those messages should be displayed.<br>
We need a way for the program to make a **decision**.<br>
That is exactly what conditional statements provide.

## 6.2 What is a Conditional Statement?

A conditional statement allows a program to execute different blocks of code depending on whether a condition is **True** or **False**.<br>
Think of it as asking a question.
```text
Is Age ≥ 18 ?
        │
   ┌────┴────┐
   │         │
 Yes         No
   │         │
 Eligible    Not Eligible
```
The program follows only **one path**.

## 6.3 Flow of a Conditional Statement<br>
Every conditional statement follows this basic pattern.
```text
             Condition
                 │
          True ? False
          │           │
          ▼           ▼
     Execute A    Execute B
          │
          ▼
         End
```
Unlike previous chapters, execution is no longer linear.


## 6.4 The `if` Statement<br>
The simplest conditional statement is `if`.<br>
**Syntax**
```python
if condition:
    statement
```
Notice two important things:
1. The condition ends with a colon (`:`).
2. The statement inside the `if` block is indented.

**Example**
```python
age = 20
if age >= 18:
    print("Eligible to Vote")
```
Output
```text
Eligible to Vote
```
**What happens internally?**

```text
age = 20
↓
20 >= 18
↓
True
↓
Execute if block
↓
Print "Eligible to Vote"
```
**If the condition is False**

```python
age = 15
if age >= 18:
    print("Eligible to Vote")
```
Output
```text
(No Output)
```
Python simply skips the block.<br>
**Visualization**

```text
age = 15
↓
15 >= 18
↓
False
↓
Skip Block
↓
Program Continues
```

## 6.5 The `if...else` Statement<br>
Sometimes we want to perform one action if the condition is true and another if it is false.<br>
Syntax
```python
if condition:
    statement
else:
    statement
```
Example
```python
age = 16
if age >= 18:
    print("Eligible")
else:
    print("Not Eligible")
```
Output
```text
Not Eligible
```
**Visualization**

```text
           Age ≥ 18 ?
          ┌─────┴─────┐
        Yes          No
         │            │
    Eligible        Not Eligible
```
Only one block executes.

## 6.6 The `elif` Statement<br>
Sometimes there are more than two possibilities.<br>
Example<br>
Marks

```text
90–100 → Grade A
75–89 → Grade B
60–74 → Grade C
Below 60 → Grade D
```
This cannot be handled efficiently using only `if` and `else`.
Python provides `elif`.<br>
Syntax

```python
if condition:
    ...
elif condition:
    ...
elif condition:
    ...
else:
    ...
```
Example

```python
marks = 82
if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 60:
    print("Grade C")
else:
    print("Grade D")
```
Output

```text
Grade B
```

## 6.7 Order Matters<br>
Python checks conditions from top to bottom.<br>
As soon as one condition is `True`, the remaining conditions are skipped.<br>
Visualization<br>
```text
marks = 82
↓
marks >= 90
False
↓
marks >= 75
True
↓
Grade B
↓
Stop Checking
```
This behavior is extremely important.

## 6.8 Nested `if` Statements<br>
An `if` statement can contain another `if`.<br>
Example
```python
age = 22
has_license = True
if age >= 18:
    if has_license:
        print("Can Drive")
```
**Visualization**
```text
Age ≥ 18 ?
      │
     Yes
      │
Has License ?
   │        │
 Yes       No
  │
Can Drive
```

## 6.9 Using Logical Operators<br>
Conditions often involve multiple checks.<br>
Example
```python
age = 25
citizen = True
if age >= 18 and citizen:
    print("Eligible to Vote")
```
Both conditions must be true.

## 6.10 Common Mistakes<br>
**Mistake 1**<br>
Forgetting the colon.
Incorrect
```python
if age >= 18
```
Correct
```python
if age >= 18:
```
**Mistake 2**<br>
Incorrect indentation.<br>
Incorrect
```python
if age >= 18:
print("Eligible")
```
Correct
```python
if age >= 18:
    print("Eligible")
```
**Mistake 3**<br>
Using `=` instead of `==`.<br>
Incorrect
```python
if age = 18:
```
Correct
```python
if age == 18:
```

## 6.11 Best Practices

* Keep conditions simple and readable.
* Use meaningful variable names.
* Avoid deeply nested conditions where possible.
* Use logical operators when appropriate.
* Maintain consistent indentation (4 spaces).
