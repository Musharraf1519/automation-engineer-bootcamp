## 1. What is Python?

Python is a high-level, interpreted, general-purpose programming language designed with one primary goal:

**Make programming simple, readable, and productive.**

Unlike many programming languages that require developers to manage memory, pointers, or complex syntax, Python allows programmers to focus on solving problems instead of struggling with the language itself.

Python follows a philosophy that emphasizes writing code that is easy to read and maintain.

Today Python is one of the most widely used programming languages in the world and powers everything from simple automation scripts to artificial intelligence systems used by companies like Google, Netflix, Microsoft, OpenAI, Amazon and Meta.

## 2. Why was Python Created?

Python was created by Guido van Rossum in 1989 and released publicly in 1991.

At that time many programming languages were powerful but difficult to write.

Guido wanted a language that was:

- Easy to read
- Easy to learn
- Easy to maintain
- Powerful enough for professional software development

Python's philosophy can be summarized in one sentence:

> Code is read far more often than it is written.

This philosophy influences almost every design decision in Python.


Compared to many other programming languages, the amount of code required is significantly less.

---


**Interpreted Language**

Python executes code through an interpreter.
This removes the need for manually compiling source code before execution.
We'll study this process in detail later in this chapter.

---

**Cross Platform**

Python programs can run on:

* Windows
* Linux
* macOS

Most programs require little or no modification.

---

**Oen Source**

Python is completely free to use.

Its source code is publicly available and maintained by a large global community.

---

**Huge Ecosystem**

Python has hundreds of thousands of libraries.

Examples include:

* NumPy
* Pandas
* Requests
* Flask
* FastAPI
* Selenium
* Playwright
* PySpark
* TensorFlow
* PyTorch

Instead of building everything from scratch, developers reuse these libraries.

## 4. Applications of Python

Python is used in almost every area of software development.

**Automation**

Examples:

* Email Automation
* File Processing
* Excel Automation
* API Automation
* Browser Automation
* PDF Processing

---

**Artificial Intelligence**

Examples:

* Chatbots
* LLM Applications
* AI Agents
* Computer Vision
* Recommendation Systems

---

**Data Engineering**

Examples:

* ETL Pipelines
* Data Cleaning
* Spark Applications

---

**Web Development**

Examples:

* REST APIs
* Authentication Systems
* Dashboards

---

**Cybersecurity**

Examples:

* Log Analysis
* Network Scanners
* Security Automation


## 5. Why Python for Automation?

Automation aims to replace repetitive manual work.<br>
Suppose a company receives hundreds of invoices every day.

Without automation:
```steps
Receive Email
↓
Download PDF
↓
Read Invoice
↓
Extract Data
↓
Update Database
↓
Send Email

```


With Python:
```steps
Receive Email
↓
Python Program
↓
Extract Invoice Data
↓
Validate Information
↓
Store Database
↓
Send Confirmation Email
```

The employee now focuses on decision making rather than repetitive work.
This is why Python has become the most popular language for automation.

## 6. Why Learn Python Before n8n?

n8n is excellent for orchestrating workflows.<br>
However, every real-world automation eventually requires custom programming.<br>
Examples include:

* Cleaning complex datasets
* Processing PDFs
* Image manipulation
* Calling multiple APIs
* AI integration
* Data transformation

In these situations Python complements n8n.<br>
Think of n8n as the workflow engine and Python as the processing engine.

# 7. How Python Executes a Program

When you run:<br>
python hello.py<br>
Python performs the following steps:

```steps
Source Code
↓
Python Interpreter
↓
Bytecode
↓
Python Virtual Machine
↓
Operating System
↓
Output
```

The interpreter reads your source code, converts it into bytecode, and the Python Virtual Machine executes that bytecode.

## 8. Installing Python

Verify installation using:
```cmd
python --version
pip --version
```

Expected Output:
```cmd
Python 3.12.x
pip 26.x.x
```

## 9. Your First Python Program

Create a file named:

01_hello_world.py

Write:

```python
print("Hello, Automation Engineer!")
```

Run:

```bash
python 01_hello_world.py
```

Output:

```
Hello, Automation Engineer!
```
