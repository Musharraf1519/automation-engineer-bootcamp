## 5.1 What is Input?<br>
Until now, every program contained values that were already written inside the code.<br>
Example:
```python
name = "Musharraf"
age = 25
```
These values never change unless the programmer edits the source code.<br>
Imagine a calculator like this.
```python
num1 = 20
num2 = 10
print(num1 + num2)
```

Every time you run it,<br>
Output
```text
30
```
What if the user wants
```
100 + 250
```
instead?<br>
The programmer shouldn't edit the source code every time.<br>
Instead, the program should ask the user.<br>
This process is called **Input**.

---

**Definition**<br>
> **Input** is the process of receiving data from a user, another program, or an external source so that it can be processed.

Examples of input:
* Keyboard
* Mouse
* File
* Database
* API
* Sensors

In this chapter, we'll focus on **keyboard input**.

## 5.2 What is Output?<br>
After processing input, a program usually displays information.<br>
Example:
```python
print("Welcome")
```
The text displayed on the screen is called **Output**.<br>
Definition:
> **Output** is the information produced by a program after processing the input.

---

**Program Flow**<br>
Every program follows the same basic cycle.
```text
            Input
              │
              ▼
        Processing
              │
              ▼
            Output
```

Example
```text
Enter First Number
↓
25
↓
Add 10
↓
35
↓
Display Result
```

## 5.4 The `input()` Function<br>
The `input()` function waits for the user to enter data.<br>
**Syntax**

```python
input("Message")
```

Example
```python
name = input("Enter your name: ")
```

Execution
```text
Enter your name:
```

User types
```text
Musharraf
```
Variable now contains
```python
name = "Musharraf"
```

---

**Internal Working**<br>
When Python reaches
```python
input()
```

it pauses execution.
```text
Program Starts
↓
input()
↓
Wait for User
↓
User Types
↓
Press Enter
↓
Program Continues
```

## 5.5 Important Observation

Everything returned by `input()` is a **string**.<br>
Example
```python
age = input("Enter Age: ")
print(type(age))
```
User enters
```text
25
```
Output
```text
<class 'str'>
```
Even though the user typed a number, Python stores it as a string.

## 5.6 Type Conversion

Suppose

```python
age = input("Enter Age: ")
```

User enters

```text
25
```

Python stores

```python
"25"
```

If you write

```python
print(age + 5)
```

Python raises an error.

Correct approach

```python
age = int(input("Enter Age: "))
```

Visualization

```text
User
↓
25
↓
input()
↓
"25"
↓
int()
↓
25
```

## 5.8 Formatted Output

Instead of

```python
print("Name:", name)
```

prefer

```python
print(f"Name : {name}")
```

Example

```python
name = "Alice"
age = 25
print(f"Name : {name}")
print(f"Age  : {age}")
```

Output

```text
Name : Alice
Age  : 25
```

## 5.10 Common Mistakes><br>
**Mistake 1**
Incorrect.
```python
age = input("Age : ")
print(age + 10)
```
Correct

```python
age = int(input("Age : "))
print(age + 10)
```

---
**Mistake 2**<br>
Using confusing prompts.<br>
Bad

```python
input()
```

Good

```python
input("Enter your Email : ")
```
