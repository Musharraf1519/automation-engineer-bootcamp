## 3.1 What is a Data Type?

Every piece of information stored in a computer has a **type**.<br>
Examples:
* `25` → Integer
* `85.5` → Floating-point number
* `"Musharraf"` → String
* `True` → Boolean

A **data type** defines:
* What kind of value is stored.
* How much memory it requires.
* What operations can be performed on it.
* How Python should interpret it.

Think of a data type as a **label** attached to every object.<br>
Example:
```text
               Object
        +------------------+
        | Value : 25       |
        | Type  : int      |
        +------------------+
```
Without knowing the type, Python wouldn't know whether:
```python
10 + 20
```
means
```
30
```
or
```
"1020"
```
The data type tells Python how to interpret the operation.

## 3.2 Why Do We Need Data Types?

Imagine a computer that stored every value without knowing its type.<br>
Suppose we have:
```python
a = 10
b = "20"
```
Now execute:
```python
print(a + b)
```
Should the result be:
```text
30
```
or
```text
1020
```

Python cannot guess your intention.<br>
That's why every object carries its own type information.<br>
When Python evaluates an operation, it first checks the data types involved and then decides whether the operation is valid.

## 3.5 Integer (`int`)

An integer represents a whole number without a decimal point.<br>
Examples:
```python
age = 25
marks = 100
temperature = -5
```

Characteristics:
* Positive or negative
* Unlimited precision (limited only by available memory)
* Immutable

Examples:
```python
print(type(25))
```

Output:
```text
<class 'int'>
```

## 3.6 Floating-Point (`float`)

A float represents numbers with decimal values.

Examples:

```python
salary = 45000.75
pi = 3.14159
```

Output:

```python
print(type(pi))
```

```text
<class 'float'>
```

## 3.7 String (`str`)

A string is a sequence of characters enclosed in quotes.
```python
name = "Musharraf"
city = 'Bangalore'
```

Strings can contain:
* Letters
* Numbers
* Symbols
* Spaces

Examples:
```python
company = "OpenAI"
phone = "9876543210"
```
Notice that a phone number is often stored as a string because we don't perform arithmetic on it.

## 3.8 Boolean (`bool`)

A Boolean has only two possible values.
```python
True
False
```

Example:
```python
is_logged_in = True
is_admin = False
```

Internally:
```python
True == 1
False == 0
```
We'll explore this further when we study conditional statements.

## 3.9 Complex Numbers (`complex`)

Python also supports complex numbers.<br>
Example:
```python
number = 4 + 3j
```

Output:
```python
print(type(number))
```
```text
<class 'complex'>
```
These are commonly used in scientific and engineering applications.

## 3.10 None (`NoneType`)<br>
Sometimes a variable intentionally has **no value**.<br>
Python represents this using `None`.<br>
Example:
```python
manager = None
```

Output:
```python
print(type(manager))
```
```text
<class 'NoneType'>
```

## 3.11 Checking Data Types

Use the built-in `type()` function.
```python
age = 25
print(type(age))
```

Output:
```text
<class 'int'>
```

## 3.12 Type Conversion<br>
Sometimes we need to convert one type into another.<br>
Example:
```python
age = "25"
print(type(age))
```

Output:

```text
<class 'str'>
```
Convert it:
```python
age = int(age)
print(type(age))
```

Output:
```text
<class 'int'>
```
Common conversions:
```python
int()
float()
str()
bool()
```

## 3.13 Common Mistakes
```python
age = "25"

print(age + 5)
```
This raises a `TypeError`.

**Correct:**
```python
age = int(age)

print(age + 5)
```
---
```python
salary = 50000
print("Salary : " + salary)
```

**Correct:**
```python
print("Salary :", salary)
```
or
```python
print(f"Salary : {salary}")
```

## 3.14 Best Practices<br>
✔ Choose the correct data type for the information.<br>
✔ Store phone numbers as strings.<br>
✔ Store monetary values as numbers.<br>
✔ Use Booleans for true/false conditions.<br>
✔ Convert data types explicitly when needed.<br>
