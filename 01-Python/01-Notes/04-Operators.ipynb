## 4.1 What is an Operator?<br>
Imagine writing the following program.
```python
age = 25
salary = 50000
```
You have successfully stored two values.<br>
Now suppose you want to increase the salary by ₹5,000.<br>
How would Python know what you mean?
```python
salary ???? 5000
```
Should it:
* Add?
* Subtract?
* Multiply?
* Divide?

Python cannot guess your intention.<br>
This is where **operators** come into the picture.<br>
An **operator** is a special symbol (or keyword) that tells Python **what operation should be performed on one or more values**.<br>
Example:
```python
salary = 50000
salary + 5000
```
Here,
```text
+
```
is the **operator**.<br>
It tells Python to perform **addition**.

---

**Visualization**

```text
          Operator
salary     +     bonus
50000      +     5000
      ↓
    55000
```
Without the operator, Python wouldn't know what to do with the two values.

## 4.2 What are Operands?<br>
Consider:
```python
10 + 20
```
Let's break it down.
```text
10      +      20
│              │
│              │
└──────┬───────┘
       │
   Operands
+
│
Operator
```
Here:
* `10` is an operand.
* `20` is an operand.
* `+` is the operator.

In general,
```text
Operand
↓
Operator
↓
Operand
```
produces a result.

## 4.5 Arithmetic Operators

Arithmetic operators perform mathematical calculations.

| Operator | Description    | Example   |
| -------- | -------------- | --------- |
| `+`      | Addition       | `10 + 5`  |
| `-`      | Subtraction    | `10 - 5`  |
| `*`      | Multiplication | `10 * 5`  |
| `/`      | Division       | `10 / 5`  |
| `//`     | Floor Division | `10 // 3` |
| `%`      | Modulus        | `10 % 3`  |
| `**`     | Exponentiation | `2 ** 3`  |

Example:

```python
a = 20
b = 6

print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a // b)
print(a % b)
print(a ** b)
```

**Understanding `/` vs `//`**<br>
This confuses many beginners.
```python
10 / 3
```
Output:
```text
3.3333333333333335
```
Python preserves the decimal part.<br>
Now,
```python
10 // 3
```
Output:
```text
3
```
Floor division removes the fractional part by rounding **down** to the nearest whole number.

---

**Understanding `%` (Modulus)**<br>
The modulus operator returns the **remainder** after division.<br>
Example:
```python
10 % 3
```
Output:
```text
1
```
Real-world uses:

* Check whether a number is even or odd.
* Circular indexing.
* Time calculations.
* Pagination.

---
**Understanding `**`**
Exponentiation raises a number to a power.
```python
2 ** 5
```
means
```text
2 × 2 × 2 × 2 × 2
```
Output
```text
32
```

## 4.6 Assignment Operators<br>
The assignment operator stores a value in a variable.
```python
salary = 50000
```
Compound assignment operators perform an operation and assignment in a single step.
```python
salary += 5000
```
Equivalent to:
```python
salary = salary + 5000
```
Common operators:
| Operator | Equivalent   |
| -------- | ------------ |
| `+=`     | `x = x + y`  |
| `-=`     | `x = x - y`  |
| `*=`     | `x = x * y`  |
| `/=`     | `x = x / y`  |
| `//=`    | `x = x // y` |
| `%=`     | `x = x % y`  |
| `**=`    | `x = x ** y` |

## 4.7 Comparison Operators<br>
Comparison operators compare two values and always return a **Boolean** (`True` or `False`).<br>
These operators are heavily used in:
* Decision making
* Validation
* Searching
* Filtering
* Loops
* Automation workflows<br>
**Available Comparison Operators**

| Operator | Meaning                  | Example    |
| -------- | ------------------------ | ---------- |
| `==`     | Equal To                 | `10 == 10` |
| `!=`     | Not Equal To             | `10 != 20` |
| `>`      | Greater Than             | `20 > 10`  |
| `<`      | Less Than                | `10 < 20`  |
| `>=`     | Greater Than or Equal To | `18 >= 18` |
| `<=`     | Less Than or Equal To    | `15 <= 20` |

**Equal To (`==`)**<br>
Checks whether two values are equal.
```python
age = 25
print(age == 25)
```
Output
```text
True
```

Visualization

```text
25 == 25
↓
Compare
↓
True
```
---

**Not Equal To (`!=`)**
```python
print(10 != 20)
```
Output
```text
True
```
---
**Greater Than (`>`)**
```python
marks = 92
print(marks > 90)
```

Output
```text
True
```

---
**Less Than (`<`)**
```python
print(10 < 5)
```

Output
```text
False
```

---
**Greater Than or Equal To (`>=`)**

```python
age = 18
print(age >= 18)
```

Output
```text
True
```
---
**Less Than or Equal To (`<=`)**

```python
temperature = 35
print(temperature <= 40)
```
Output
```text
True
```
---
**Think Before You Run**

Predict the output.
```python
print(15 >= 20)
```
Don't run it immediately.

## 4.8 Logical Operators<br>
Logical operators combine multiple conditions.<br>
Imagine an ATM.<br>

A customer can withdraw money only if:
* Card is valid
* PIN is correct

Both conditions must be true.<br>
This is where logical operators are used.
---

**Types**

| Operator | Meaning                             |
| -------- | ----------------------------------- |
| `and`    | Both conditions must be True        |
| `or`     | At least one condition must be True |
| `not`    | Reverses the result                 |

---

**AND**

```python
is_logged_in = True
is_admin = True

print(is_logged_in and is_admin)
```
Output
```text
True
```

Visualization
```text
True
AND
True
↓
True
```
---

**OR**

```python
print(True or False)
```

Output
```text
True
```
Only one condition needs to be true.

---

**NOT**
```python
print(not True)
```
Output
```text
False
```
NOT simply reverses the Boolean value.

**Real-World Example**
```python
age = 22
has_license = True
print(age >= 18 and has_license)
```
Output
```text
True
```

## 4.9 Identity Operators<br>
Identity operators compare **whether two variables refer to the same object**, not whether they have the same value.<br>
This concept connects directly to Chapter 2 (Variables & Memory).

**Operators**

| Operator | Meaning           |
| -------- | ----------------- |
| `is`     | Same object       |
| `is not` | Different objects |

Example
```python
x = 100
y = x
print(x is y)
```

Visualization
```text
        Memory
+----------------+
| Integer 100    |
+----------------+
      ▲     ▲
      │     │
      x     y
```

Both variables point to the same object.<br>
Output
```text
True
```
---
Value vs Identity
```python
a = [1, 2]
b = [1, 2]
```
```python
print(a == b)
```
Output
```text
True
```
because the values are equal.
Now,
```python
print(a is b)
```
Output
```text
False
```
because they are different objects in memory.

## 4.12 Operator Precedence<br>
Sometimes an expression contains multiple operators.<br>
Example
```python
10 + 5 * 2
```
Which operation happens first?<br>
Addition?<br>
Multiplication?<br>
Python follows precedence rules.<br>
Visualization
```text
10 + 5 × 2
↓
10 + 10
↓
20
```
Multiplication happens before addition.

---

**Common Precedence Order**

| Priority | Operator   |
| -------- | ---------- |
| Highest  | `()`       |
|          | `**`       |
|          | `* / // %` |
|          | `+ -`      |
|          | Comparison |
|          | `not`      |
|          | `and`      |
| Lowest   | `or`       |

## 4.13 Best Practices<br>
✔ Use parentheses to improve readability.
```python
total = (price + tax) * quantity
```
✔ Don't write overly complex expressions.<br>
✔ Use `==` for value comparison.<br>
✔ Use `is` only when checking object identity or `None`.
```python
if manager is None:
    ...
```
✔ Use meaningful variable names.

## 4.14 Common Mistakes<br>
**Mistake 1**
```python
if age = 18:
```
Incorrect.

Correct
```python
if age == 18:
```

---
**Mistake 2**

Using `is` instead of `==`<br>
Incorrect
```python
name1 = "John"
name2 = "John"
print(name1 is name2)
```
Use
```python
print(name1 == name2)
```
unless you're intentionally checking object identity.

---
**Mistake 3**<br>
Ignoring precedence.<br>
Instead of
```python
total = price + tax * quantity
```
write
```python
total = (price + tax) * quantity
```
if that matches your intended calculation.
