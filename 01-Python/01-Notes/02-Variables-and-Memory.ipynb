## 2.1 Why Do Variables Exist?

Imagine writing a payroll application.<br>
Without variables:

```python
print("Musharraf")
print(45000)
print("Software Engineer")
```

Looks fine.<br>
Now imagine 50,000 employees.<br>

This is why variables exist.<br>
Variables allow us to **name data** so we can use it repeatedly.

## 2.2 What is a Variable?

Most books say:

> A variable is a container that stores data.

This explanation is **not entirely correct** for Python.<br>
A better definition is:

> A variable is a **name (reference)** that points to an object stored in memory.

This distinction is one of the most important concepts in Python.

## 2.3 Variables vs Objects

One of the biggest misconceptions beginners have is believing that a variable is a box that stores a value.<br>
Consider the following statement:
```python
age = 25
```
Many beginners imagine the computer doing something like this:
```text
+----------------+
| Variable : age |
+----------------+
|       25       |
+----------------+
```

According to this model, the variable itself contains the value.<br>
Although this mental model works for simple programs, **it is not how Python actually stores data.**

---

### What Actually Happens?
Python treats **everything as an object**.
When the interpreter executes
```python
age = 25
```
it performs two separate operations.<br>
**Step 1 — Create an Integer Object**<br>
Python first creates an integer object representing the value **25** somewhere in memory.
```text
Memory
+----------------------+
| Integer Object       |
|----------------------|
| Value : 25           |
| Type  : int          |
+----------------------+
```

Notice something important.<br>
At this stage, there is **no variable named `age` yet.**<br>
Python has only created an object.<br>
**Step 2 — Create a Reference**<br>
Now Python creates the variable name **age**.<br>
However, instead of storing the value inside the variable, Python stores a **reference** (or address) that points to the integer object.
```text
               Memory
      +----------------------+
      | Integer Object       |
      |----------------------|
      | Value : 25           |
      | Type  : int          |
      +----------------------+
                ▲
                │
                │
             Reference
                │
                │
              age
```

The variable does **not** contain the integer.<br>
The variable only remembers **where the integer object lives in memory.**

---

### Why Does Python Work This Way?

Imagine a very large object.

```python
employee = {
    "id":101,
    "name":"John",
    "salary":85000,
    "department":"IT",
    ...
}
```

Suppose this object occupies **5 MB** in memory.<br>
Now imagine every time you wrote
```python
manager = employee
```
Python copied all 5 MB again.<br>
Memory would quickly become exhausted.<br>
Instead, Python simply creates another reference.
```text
                    Memory
          +---------------------------+
          | Employee Object           |
          | Size : 5 MB               |
          +---------------------------+
               ▲                 ▲
               │                 │
            employee         manager
```
Both variables refer to the **same object**.<br>
No duplication occurs.<br>
This makes Python much more memory efficient.<br>
always read it mentally as
> Create (or locate) an integer object representing 25 and make the name **age** refer to that object.
This mental model will help you understand assignment, copying, mutability, functions, classes, and object-oriented programming later in the course.


## 2.4 Objects in Python
Everything in Python is an object.<br>
Examples:

```python
10
10.5
"Hello"
True
[1,2,3]
{"name":"John"}
```
All of these are objects.<br>
Python creates them in memory.<br>
Variables simply reference them.

## 2.5 Creating a Variable

```python
name = "Musharraf"
age = 25
salary = 75000.50
```

Syntax
```text
variable_name = value
```

## 2.6 Variable Assignment

Example

```python
country = "India"
```

Python performs roughly these steps:

1. Create string object.
2. Allocate memory.
3. Store string.
4. Make variable point to it.

Visualization

```text
country
      │
      ▼
+-----------------+
| "India"         |
+-----------------+
```

## 2.7 Multiple Variables

```python
a = 10
b = 20
c = 30
```

## 2.8 Multiple Assignment

```python
a, b, c = 10, 20, 30
```

Python automatically assigns

```text
a → 10
b → 20
c → 30
```

## 2.9 Same Value Assignment

```python
x = y = z = 100
```

All three variables reference the same integer object.<br>
Visualization

```text
x ─┐
y ─┼────► 100
z ─┘
```

## 2.10 Swapping Variables
Most languages require
```text
temp = a
a = b
b = temp
```

Python supports

```python
a = 10
b = 20
a, b = b, a
```
No temporary variable required.

## 2.11 Dynamic Typing
Python does **not** require explicit datatype declarations.
```python
age = 25
```
Later
```python
age = "Twenty Five"
```
This is perfectly valid.<br>
The variable now references a string object instead of an integer object.

## 2.12 Checking Memory Address
Python provides
```python
id()
```

Example

```python
a = 100
print(id(a))
```

Output
```
235423423432
```

(The actual value will differ.)
`id()` returns the identity of the object during its lifetime.

## 2.14 Variable Naming Rules

**Valid**

```python
student_name = "John"
salary2 = 50000
_marks = 95
```

**Invalid**

```python
2salary = 50000
student-name = "John"
class = 20
```

## 2.15 Naming Conventions

**Good**
```python
employee_salary
student_name
customer_id
```

**Bad**

```python
x
abc
temp12345
aaaa
```

A good variable name describes the data it references.


## 2.16 Constants
Python has no true constants.<br>
Convention:
```python
PI = 3.14159
MAX_RETRY = 5
```
Uppercase indicates the value should not be modified.

## 2.17 Best Practices
✔ Use meaningful names.
✔ Keep names short but descriptive.
✔ Follow `snake_case`.
✔ Avoid single-letter variables except in loops or mathematics.
✔ Avoid reserved keywords.
✔ Don't abbreviate unnecessarily.

## 2.18 Common Beginner Mistakes
```python
Name = "John"
name = "Alex"
```
Python treats these as different variables.

---
```python
salary = "50000"
salary + 1000
```
This raises an error because `"50000"` is a string.

---
```python
x = 10
x = "Hello"
```
Legal, but avoid changing types without a good reason—it makes code harder to understand.
